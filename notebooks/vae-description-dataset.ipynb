```python
import os
import pandas as pd
from pathlib import Path
from datasets import load_dataset
```

```python
path = "../outputs/caption_inference/2025-12-09_11-46-14"

df_train = pd.read_csv(Path(path) / "train_predictions.csv")
df_val = pd.read_csv(Path(path) / "validation_predictions.csv")
df_test = pd.read_csv(Path(path) / "test_predictions.csv")
```

```python
# Cut prompt from captions
def cut_prompt(caption):
    prompt_end = "Use all the tags provided to create a comprehensive and engaging description."
    caption = caption.split(prompt_end)[-1].strip()
    caption = caption.split("\n")[0].strip()
    return caption
```

```python
df_train['prediction'] = df_train['prediction'].apply(cut_prompt)
df_val['prediction'] = df_val['prediction'].apply(cut_prompt)
df_test['prediction'] = df_test['prediction'].apply(cut_prompt)

df_train.rename(columns={"prediction": "caption"}, inplace=True)
df_val.rename(columns={"prediction": "caption"}, inplace=True)
df_test.rename(columns={"prediction": "caption"}, inplace=True)
```

```python
output_path = "../data/vae_tags_description_dataset"
os.makedirs(output_path, exist_ok=True)

df_train.to_csv(Path(output_path) / "train.csv", index=False)
df_val.to_csv(Path(output_path) / "validation.csv", index=False)
df_test.to_csv(Path(output_path) / "test.csv", index=False)
```

```python

data_files = {
    "train": str(Path(output_path) / "train.csv"),
    "validation": str(Path(output_path) / "validation.csv"),
    "test": str(Path(output_path) / "test.csv"),
}
dataset = load_dataset("csv", data_files=data_files)
dataset.push_to_hub("bsienkiewicz/musicgen-vae-tags-description-dataset")
```
